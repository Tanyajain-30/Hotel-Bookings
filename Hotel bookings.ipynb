# Checking Types of Hotels

```python
plt.figure(figsize=(6,3))
sns.countplot(x="hotel", data=df, width=0.5, stat='count')
plt.title("Types of hotel")
plt.show()
```

As we can in above plot most of the bookings done in city hotels

# Reservation Status

```python
df['reservation_status'].value_counts()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x="reservation_status", data=df, width=0.5, hue='hotel')
```

As we can see in above plot most of the bookings canceled in city hotel as compair to resort

# Guest Status

```python
df['Guest Type'].value_counts()
```

```python
x=df['Guest Type'].value_counts().index
y=df['Guest Type'].value_counts().values
plt.pie(y, autopct="%0.2f%%", labels=x, startangle=45)
plt.legend()
plt.show()
```

As we can see in above chart, most of the bookings done by couples

# Checking Reservetion Status According To Guest Status

```python
plt.figure(figsize=(6,4))
sns.countplot(x="reservation_status", data=df, width=0.5, hue='Guest Type')
plt.show()
```
